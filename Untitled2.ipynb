```python
import pandas as pd
data = pd.read_csv('goodreads_data.csv')
import matplotlib.pyplot as plt

# Count the number of books contributed by each author
author_counts = data['Author'].value_counts().head(10)

# Plot a pie chart
plt.figure(figsize=(8, 8))
author_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('Top 10 Authors and Their Book Contributions')
plt.ylabel('')
plt.show()
```

```python
# Plot a histogram of average ratings
plt.figure(figsize=(8, 6))
plt.hist(data['Avg_Rating'], bins=20)
plt.title('Distribution of Average Ratings')
plt.xlabel('Average Rating')
plt.ylabel('Count')
plt.show()

```

```python
# Create a scatter plot of average rating vs. number of ratings
plt.figure(figsize=(8, 6))
plt.scatter(data['Num_Ratings'], data['Avg_Rating'])
plt.title('Average Rating vs. Number of Ratings')
plt.xlabel('Number of Ratings')
plt.ylabel('Average Rating')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count the number of books in each genre
genre_counts = data['Genres'].value_counts()

# Plot a bar chart
plt.figure(figsize=(12, 8))
plt.gca().set_facecolor('white')  # Set background color

# Customize color palette
colors = plt.cm.Set3(range(len(genre_counts)))

# Plot the bar chart
genre_counts.plot(kind='bar', color=colors)

# Add grid lines
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Customize font size
plt.title('Distribution of Books by Genre', fontsize=16)
plt.xlabel('Genre', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45, fontsize=10)

# Remove unnecessary elements
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tick_params(axis='both', which='both', bottom=False, left=False)

plt.show()

#  visualizes the distribution of books by genre.
```

```python
import matplotlib.pyplot as plt

outlets = ["NKSK", "Fertilizer Dealers", "City Dealers", "IP Dealers", "Central Police Canteens",
           "Kendriya Bhandar", "Big Bazaar", "Star Bazaar", "GNFC Neem Shops", "Ecommerce"]
sales_percentages = [23.7, 33.1, 20.0, 4.8, 8.6, 3.6, 2.7, 1.5, 1.8, 0.2]

plt.figure(figsize=(10, 6))
plt.bar(outlets, sales_percentages)

plt.title("Sale of Neem Products via Retail Outlets (until November 2017)")
plt.xlabel("Retail Outlets")
plt.ylabel("Sales (%)")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
